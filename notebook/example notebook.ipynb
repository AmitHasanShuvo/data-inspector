## Importing data inspector 

```python
import data_inspector as di
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

### Line plot : line_plot(data, x_data, y_data, x_label="", y_label="", title="")

```python
di.line_plot(data, 'PassengerId', 'Age', x_label="PassengerId", y_label="Age", title="Line Plot")
```

## Skew feature: plot_skewed_feature(data, column)

```python
di.plot_skewed_feature(data, 'Age')
```

## Showing data distribution: show_distribution(data, column)

```python
di.show_distribution(data, 'Age')
```

## Scatter plot: plot_scatter(data,x_data, y_data)

```python
di.plot_scatter(data,'Age', 'Pclass')
```

## Correlation plot: plot_correlation(data)

```python
di.plot_correlation(data)
```

## Create histogram: histogram(data,column, x_label, y_label, title)

```python
di.histogram(data,'Age', 'X axis', 'Y axis', 'Histogram')
```

## Create bar plot: plot_bar(data, column, xlabel, ylabel, title)

```python
di.plot_bar(data, 'Sex', 'X axis', 'Y axis', 'Bar Plot')
```

## Create boxplots of all features: box_plot(data)

```python
di.box_plot(data)
```

## Checking dataset's shape: datasetShape(data)

```python
di.datasetShape(data)
```

## Get dataset's diagnostic plots: diagnostic_plots(data, variable)

```python
di.diagnostic_plots(data, 'Age')
```

## Divide numerical and categorical features: divideFeatures(data)

```python
di.divideFeatures(data)
```

## Fill NaN values: fillNan(data, column, value)

```python
di.fillNan(data, 'Cabin', 'Missing')
```

## Plotting kde plots: plot_cont_kde(data, var)

```python
di.plot_cont_kde(data, 'Age')
```
